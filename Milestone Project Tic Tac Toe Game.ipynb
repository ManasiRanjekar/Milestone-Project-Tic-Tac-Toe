# Tic Tac Toe Game-

```python
from IPython.display import clear_output
```

```python
#the board itself!
board=['#','X','O','X','O','X','O','X','O','X']
```

```python
#This function is used to print out the board.
def display_board(board):
    #Clear_output allows us to show only one version of board,because we don't want multiple boards to be displayed.
    clear_output()
    print(board[1] + '|' + board[2] + '|' + board[3])
    print('-+-+-')
    print(board[4] + '|' + board[5] + '|' + board[6])
    print('-+-+-')
    print(board[7] + '|' + board[8] + '|' + board[9])
```

```python
#Take input from the player and assign the marker 'X' or 'O'.
def player_input():
    #variable to receive player input.
    marker = ""
    #while loop until we have the correct answer.
    while not (marker == "X" or marker == "O"):
        marker = input("Player 1 - Please pick a marker 'X' or 'O': ").upper()
    #player 2 don't have any options here.
    if marker == "X":
        return ('X','O')
    else:
        return ('O','X')
```

```python
#the place_marker function takes in three parameteres-the board list,marker and position and then assigns the desired position on the board.
def place_marker(board,marker,position):
    board[position]= marker
```

```python
#win check function checks if all win combinations share the same marks-all rows,allcolumns and two diagonals
def win_check(board, mark):
    #check rows
    if board[1] == board[2] == board[3] == mark:
        return True
    if board[4] == board[5] == board[6] == mark:
        return True
    if board[7] == board[8] == board[9] == mark:     
        return True
    #check columns
    if board[1] == board[4] == board[7] == mark:
        return True
    if board[2] == board[5] == board[8] == mark:
        return True
    if board[3] == board[6] == board[9] == mark:
        return True
    #check diagonally
    if board[1] == board[5] == board[9] == mark:
        return True
    if board[3] == board[5] == board[7] == mark:
        return True
    return False  
```

```python
# the random library decides randomly which player will go first.
import random
def choose_first():
    if random.randint(0,1) == 0:
        return "Player 2"
    else:
        return "Player 1 "
```

```python
#check if the board space is marked or not.
def space_check(board,position):
    #a non marked space should be equal to " ".
    return board[position] == " "
#checking if the board is full.
#If board is full return true,otherwise return false.
def full_board_check(board):
    for i in range(1,10):
        if space_check(board, i):
            #call this function to return False for an empty space on board
            return False
    #returns true if board is full
    return True 
```

```python
# a function that asks the player his next move (1-9) and use space_check to check whether the space is free or not.
def player_choice(board):
    #set a placeholder
    position = 0
    #while loop to check if the number is on the board and if the space is still available
    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board,position):
        position = int(input('Choose a position (1-9): '))
    return position
```

```python
#asks if the players want to play again.Boolean value true for yes.
def replay():
    choice = input("Do you want to play again ?Yes or No" ).lower()
    if choice == 'yes':
        return True
    if choice =='no':
        return False
```

```python
print ('Welcome to Tic Tac Toe!')
# Keep the game running through a while loop
while True:
    # Setup the board with empty strings...
    # ...because those are expected in display_board()
    the_board = [' ']*10
    # Call player interaction
    player1_marker, player2_marker = player_input()
    # Let the CPU choose who goes first
    turn = choose_first()
    print(turn + ' will go first.')
    # User interaction to start the game
    play_game = input('Ready? Enter Yes or No')
    if play_game.lower()[0] == 'yes' or play_game.lower()[0] == 'y':
        game_on=True
    else:
        game_on=False
    # Assuming True the gameplay starts
    while game_on:
        # Player's 1 turn
        if turn == 'Player 1':
            # Show the board to players
            display_board(the_board)
            # Allow player 1 to choose a position
            position = player_choice(the_board)
            # Place the icon on position assigned
            place_marker(the_board, player1_marker, position)
            # Check win condition
            if win_check(the_board, player1_marker):
                display_board(the_board)
                print ('Congratulations!Player 1, won the game!')                
                # This will break out of the loop
                game_on = False
            # checking for a tie
            else:
                if full_board_check(the_board):
                    display_board(the_board)
                    print("It's a draw!!")
                    game_on = False
                else:
                    turn = 'Player 2'
        #same process for player 2
        else:
            #Player's 2 turn
            display_board(the_board)
            position = player_choice(the_board)
            place_marker(the_board, player2_marker, position)
            if win_check(the_board, player2_marker):
                display_board(the_board)
                print ('Congratulations!Player 2, won the game!')
                game_on = False
            else:
                if full_board_check(the_board):
                    display_board(the_board)
                    print("It's a draw!!")
                    game_on = False
                else:
                    turn = 'Player 1'
    if not replay():
        break
```
